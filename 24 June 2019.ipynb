```python
# Function to generate marks data for n students

from random import randint

def generateMark(n, l, u):
    filepath = 'DataFiles/marks.txt'
    with open('filepath', 'w') as f:
        for i in range(0, n):
            r=randint(l, u)
            f.write(str(r)+'\n')
    return
generateMark(50,1,100)
```
